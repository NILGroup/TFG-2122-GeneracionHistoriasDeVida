```python
import matplotlib.pyplot as plt
from scipy.ndimage.filters import gaussian_filter1d
```

```python
def grafica(titulo, valores1, valores2, conjunto1, conjunto2,highest=0,figsize=(12,8)):

  steps1 = [i*500 for i in range(highest,highest+len(valores1))]
  steps2 = [i*500 for i in range(highest,highest+len(valores2))]

  # bspline = interpolate.make_interp_spline(steps,loss_per_100_steps_total, k=3)
  # y_new = bspline(steps)
  # plt.plot(steps, y_new)

  plt.figure(figsize=figsize)


  ysmoothed1 = gaussian_filter1d(valores1, sigma=1.1)
  line1 = plt.plot(steps1, ysmoothed1, linestyle='--', marker='o',label=conjunto1)

  ysmoothed2 = gaussian_filter1d(valores2, sigma=1.1)
  line2 = plt.plot(steps2, ysmoothed2, linestyle='--', marker='o',label=conjunto2)


  
  plt.axvline(0)
  plt.text(200,highest,'epoch 1',rotation=0,fontsize = 15)
  plt.axvline(length[0]*500)
  plt.text(3700,highest,'epoch 2',rotation=0,fontsize = 15)
  plt.axvline(length[1]*500)
  plt.text(7200,highest,'epoch 3',rotation=0,fontsize = 15)
  #plt.legend(prop={'size': 20})


  #steps = [i for i in range(len(val_per_1000_steps))]
  #plt.plot(steps, val_per_1000_steps)
  plt.title(titulo,fontsize = 15)
  plt.xlabel('Steps',fontsize = 15)
  plt.ylabel(titulo,fontsize = 15)

  
  plt.show()
```

```python
grafica("loss", losstrain_per_steps, lossval_per_steps, "train","test")
```

```python
l1 = losstrain_per_steps[3:]
l2 = lossval_per_steps[3:]
grafica("loss", l1 , l2, "train","test", 3,(6, 4))
```

```python
grafica("bleu", bleutrain_per_steps, bleuval_per_steps, "train","test",0.48)
```

## Loss a través de epochs

```python
with open ('Modelo T5 Final/Modelos/Loss Entrenamiento/lossTotal.ob', 'rb') as fp:
    loss_per_100_steps_total = pickle.load(fp)
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [10, 6]
plt.rcParams["figure.autolayout"] = True

loss_per_100_steps_1epoch = loss_per_100_steps_total[:int(len(loss_per_100_steps_total)/5)]

steps = [i*100 for i in range(len(loss_per_100_steps_1epoch))]



# bspline = interpolate.make_interp_spline(steps,loss_per_100_steps_total, k=3)
# y_new = bspline(steps)
# plt.plot(steps, y_new)

from scipy.ndimage.filters import gaussian_filter1d

ysmoothed = gaussian_filter1d(loss_per_100_steps_1epoch, sigma=1.5)
plt.plot(steps, ysmoothed)


#steps = [i for i in range(len(val_per_1000_steps))]
#plt.plot(steps, val_per_1000_steps)
plt.title('Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.show()
```
